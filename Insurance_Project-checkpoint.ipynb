## Data Cleaning

```python
df_train = pd.read_csv('QUB_Insurance_Data_Assignment_Training.csv')
df_scoring = pd.read_csv('QUB_Insurance_Data_Assignment_Scoring.csv')

# Remove outliers for training dataset
Q1 = df_train.quantile(0.25)
Q3 = df_train.quantile(0.75)
IQR = Q3-Q1
print(IQR)
df_train = df_train[~((df_train < (Q1 - 1.5*IQR)) | (df_train > (Q3+1.5*IQR))).any(axis=1)]
df_train['MotorValue'] = df_train['MotorValue'].clip(lower=0)

# Remove outliers for predicting dataset
Q1 = df_scoring.quantile(0.25)
Q3 = df_scoring.quantile(0.75)
IQR = Q3-Q1
print(IQR)
df_scoring = df_scoring[~((df_scoring < (Q1 - 1.5*IQR)) | (df_scoring > (Q3+1.5*IQR))).any(axis=1)]
df_scoring['MotorValue'] = df_scoring['MotorValue'].clip(lower=0)
    
df_train.describe()
```

```python
datacleaner = [df_train, df_scoring]
for dataset in datacleaner:
    #Remove occupation
    dataset.drop(['Occupation'], axis=1, inplace=True)
    #Merge values
    dataset['Gender'].replace(['f', 'm'],['female', 'male'],inplace=True)
    #Replace null values
    dataset['CreditCardType'].fillna('NoCreditCard', inplace=True)
    dataset['MotorValue'].fillna(dataset['MotorValue'].median(), inplace=True)
    dataset['MotorType'].fillna('NoMotorType', inplace=True)
    dataset['HealthType'].fillna('NoHealthType', inplace=True)
    dataset['TravelType'].fillna('NoTravelType', inplace=True)
    dataset['HealthDependentsAdults'].fillna(0, inplace=True)
    dataset['HealthDependentsKids'].fillna(0, inplace=True)
    #Change float to int
    dataset.HealthDependentsAdults = dataset.HealthDependentsAdults.astype(int)
    dataset.HealthDependentsKids = dataset.HealthDependentsKids.astype(int)

#Merge values    
df_train['PrefChannel'].replace(['E', 'P', 'S'],['Email', 'Phone','SMS'],inplace=True)
```

```python
#4.Encoding
def data_encoding(dataset):
    X = dataset.iloc[:,[1] + [i for i in range(5,18)]]
    y = dataset.iloc[:,-1]
    # Encoder
    oe = OrdinalEncoder()
    lb = LabelEncoder()
    oh = OneHotEncoder()
    mms = MinMaxScaler()
    pw = PowerTransformer(method='yeo-johnson')
    # Encode categorical features by onehot encoding
    # scale numerical features by minmax scaler
    # Make numerical features more gaussian by yeo-johnson
    categorical_col = X.select_dtypes(include=['object']).columns
    numerical_col = X.select_dtypes(include=['int64', 'float64']).columns
    t = [('cat', oh, categorical_col), ('num', mms, numerical_col), ('tran', pw, numerical_col)]
    transformer = ColumnTransformer(transformers=t, remainder='passthrough')
    #Input features encoding
    X = transformer.fit_transform(X)
    #Target label encoding
    y = lb.fit_transform(y)
    return X, y, transformer
    
```

```python
X, y, transformer = data_encoding(df_train)

print(y[:5])
```

```python
models = [
    #Ensemble Methods: http://scikit-learn.org/stable/modules/ensemble.html
    ('ada', AdaBoostClassifier()),
    ('bc', BaggingClassifier()),
    ('etc', ExtraTreesClassifier()),
    ('gbc', GradientBoostingClassifier()),
    ('rfc', RandomForestClassifier()),
    ('knn', KNeighborsClassifier()),
    ('svc', SVC(probability=True)),
    ('xgb', XGBClassifier()),
]

# Variables for average classification report
originalclass = []
predictedclass = []
#Make our customer score
def classification_report_with_accuracy_score(y_true, y_pred):
    originalclass.extend(y_true)
    predictedclass.extend(y_pred)
    return accuracy_score(y_true, y_pred) # return accuracy score

for m in models:
    print('-'*20,'\n', m[0])
    #pipeline = Pipeline(steps=[('prep', transformer),m])
    cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
    n_scores = cross_val_score(m[1], X, y, cv=cv, \
                               scoring=make_scorer(classification_report_with_accuracy_score), n_jobs=1)
    #print('Accuracy: %.3f (%.3f)' %(np.mean(n_scores), np.std(n_scores)))
    print(classification_report(originalclass, predictedclass)) 
    print(n_scores.mean())
```

```python
C=list(np.arange(0.01,1,100))
gamma=list(np.arange(0.0001,0.001,10))
kernel=['rbf', 'poly', 'linear','sigmoid']
hyper={'kernel':kernel,'C':C}
gd=GridSearchCV(estimator=SVC(),param_grid=hyper,verbose=True)
gd.fit(X,y)
print(gd.best_score_)
print(gd.best_estimator_)
```

```python
# choose ada, gbc, rfc, svc
# optimize hyperparameter
X, y, transformer = data_encoding(df_train)

#ada
n_estimators=list(range(600,1000,100))
learn_rate=list(np.arange(0.005,0.02,10))
hyper={'n_estimators':n_estimators,'learning_rate':learn_rate}
gd=GridSearchCV(estimator=AdaBoostClassifier(),param_grid=hyper,verbose=True)
gd.fit(X,y)
print(gd.best_score_)
print(gd.best_estimator_)
#gbc
n_estimators=list(range(100,1000,100))
learn_rate=list(np.arange(0.01,0.05,10))
hyper={'n_estimators':n_estimators,'learning_rate':learn_rate}
gd=GridSearchCV(estimator=GradientBoostingClassifier(),param_grid=hyper,verbose=True)
gd.fit(X,y)
print(gd.best_score_)
print(gd.best_estimator_)

#SVC
C=list(np.arange(0.4,0.7,10))
gamma=list(np.arange(0.01,0.1,10))
kernel=['rbf','linear', 'poly', 'sigmoid']
hyper={'kernel':kernel,'C':C,'gamma':gamma}
gd=GridSearchCV(estimator=SVC(),param_grid=hyper,verbose=True)
gd.fit(X,y)
print(gd.best_score_)
print(gd.best_estimator_)

#xgb
max_depth = list(range(1,6))
min_child_weight = list(range(1,5))
n_estimators = list(range(80,160,10))
hyper={'max_depth':max_depth, 'min_child_weight':min_child_weight, 'n_estimators':n_estimators}
gd=GridSearchCV(estimator=XGBClassifier(random_state=0),param_grid=hyper,verbose=True)
gd.fit(X,y)
print(gd.best_score_)
print(gd.best_estimator_)

#rfc
n_estimators=range(150,200,10)
hyper={'n_estimators':n_estimators}
gd=GridSearchCV(estimator=RandomForestClassifier(random_state=0),param_grid=hyper,verbose=True)
gd.fit(X,y)
print(gd.best_score_)
print(gd.best_estimator_)

```

```python
#ensembing
estimators=[('SVC',SVC()),
            ('ADA', AdaBoostClassifier(learning_rate=0.005, n_estimators=900)),
            ('GBC', GradientBoostingClassifier(learning_rate=0.01, n_estimators=900)),
            ('XGB', XGBClassifier(max_depth=2,min_child_weight=2,n_estimators=80)),
            ('ETC',ExtraTreesClassifier\
            (bootstrap=True, criterion="entropy", max_features=0.8, min_samples_leaf=19, min_samples_split=20, n_estimators=100))
]

ensemble_lin_rbf=VotingClassifier(estimators=estimators)

# Variables for average classification report
originalclass = []
predictedclass = []
#Make our customer score
def classification_report_with_accuracy_score(y_true, y_pred):
    originalclass.extend(y_true)
    predictedclass.extend(y_pred)
    return accuracy_score(y_true, y_pred) # return accuracy score

cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
n_scores = cross_val_score(ensemble_lin_rbf, X, y, cv=cv, \
                           scoring=make_scorer(classification_report_with_accuracy_score), n_jobs=1)
print(classification_report(originalclass, predictedclass)) 
print('Accuracy: %.3f (%.3f)' %(np.mean(n_scores), np.std(n_scores)))
```
